```python
data_sample=data.sample(100,random_state=42)
data_sample.head()
#get sample from data
```

```python
data_pop_age_mean=data['age'].mean()
print('age pop mean= ',data_pop_age_mean)
data_sample_age_mean=data_sample['age'].mean()
data_sample_age_std=data_sample['age'].std()
print('age sample mean= ',data_sample_age_mean)
# calculate population and sample mean for age
```

```python
confidence_level = 0.95
degrees_freedom = len(data_sample) - 1
alpha = 1 - confidence_level
t_critical = t.ppf(1 - alpha/2, degrees_freedom)
margin_error = t_critical * (data_sample_age_std / np.sqrt(len(data_sample)))
confidence_interval = (data_sample_age_mean - margin_error, data_sample_age_mean + margin_error)
print("95% Confidence Interval:", confidence_interval)
# calaculate interval estimation
```

```python
from scipy.stats import ttest_1samp
t_statistic, p_value = ttest_1samp(data_sample['age'], popmean=data_pop_age_mean)
print("T-statistic:", t_statistic)
print("p-value:", p_value)
#calculate T-statistic and p-value
#reject HO and accept Ha
reject = p_value < 0.05
print("Reject HO:", reject)
#reject HO m1=m2
#accept m1!=m2
```

```python
from scipy.stats import chi2

sample_variance = np.var(data_sample['balance'], ddof=1)
sigma2_0 = np.var(data['age'], ddof=1)
sample_size = len(data['balance'])

chi2_stat = (sample_size - 1) * sample_variance / sigma2_0

df = sample_size - 1

p_value = chi2.sf(chi2_stat, df) * 2

print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)

#calculate T-statistic and p-value
#reject HO and accept Ha
#reject HO m1=m2
#accept m1!=m2
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(data['age'], data['balance'], equal_var=True)


t_stat_welch, p_value_welch = ttest_ind(data['age'], data['balance'], equal_var=False)

print("Two-sample t-test (equal variances) - t-statistic:", t_stat, ", p-value:", p_value)
print("Welch's t-test (unequal variances) - t-statistic:", t_stat_welch, ", p-value:", p_value_welch)
#calculate T-statistic and p-value
#reject HO and accept Ha
#reject HO m1=m2
#accept m1!=m2
```

```python
from scipy.stats import f_oneway
f_statistic, p_value = f_oneway(data['age'], data['balance'],data['campaign'])
print("F-statistic:", f_statistic)
print("p-value:", p_value)
#calculate T-statistic and p-value
#reject HO and accept Ha
#reject HO m1=m2
#accept m1!=m2
```
